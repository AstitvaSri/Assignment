# Monte Carlo Search Tree 

Select => Expand => Simulate => Back Up


- Select the first child using a policy based on experience
    eg, we go 4/7 => 2/3 => 1/2 => 1/1

- Now we reach the end of the tree so we do an expansion

- To do the next step we first assume an action and then simulate the next possible state.
Now depending on the number of actions we will take some random number of states. To chose a new state you apply heuristics fot better approximation.

- After you are tired searching you update the whole tree, just like in a segment tree or min-max

#### Hence we are using reinforcement learning on the policy. Hence we do a policy evaluation, simulaate,reward  and then keep on searching 

## Properties of monte carlo search tree 
1. Useful for large states 
2. Needs alot of states
3. Planning time is independant of no. of |s| that we have already take place 
4. Running time is exponential in the effective horizon for O(|A|*X)^H)
    1. Horizon => How far in event space do we look for 
    2. |A| action space 
    3. X

# Audio Convolution 

```matlab
audio_file = audiorecorder(44100,24,1,1)
```

```matlab
disp('Start speaking.')
recordblocking(audio_file, 5);
disp('End of Recording.');
```

```matlab
play(audio_file);
```

```matlab
y = getaudiodata(audio_file);
plot(y);
```

```matlab
[P,Q] = rat(24/44.1);
y24 = resample(y,P,Q);
P24 = audioplayer(y24,24000);
play(P24)
```

```matlab
[P,Q] = rat(16/44.1);
y16 = resample(y,P,Q);
P16 = audioplayer(y16,16000);
play(P16)
```

```matlab
[P,Q] = rat(8/44.1);
y8 = resample(y,P,Q);
P8 = audioplayer(y8,8000);
play(P8)
```

```matlab
[P,Q] = rat(4/44.1);
y4 = resample(y,P,Q);
P4 = audioplayer(y4,4000);
play(P4)
```

```matlab
church_bullet = audioread('church_bullet.wav');
church = audioplayer(church_bullet,44100);
play(church);
plot(church_bullet);
```

```matlab
church_y = conv(y,church_bullet(:,1));
church_conv_audio = audioplayer(church_y,44100);
plot(church_y)
play(church_conv_audio)
```
